```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('MBA.csv')
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
df['admission']= df['admission'].fillna('Not Confirmed')
df['race']=df['race'].fillna('Unknown')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
df.drop('application_id', axis=1, inplace=True)
```

# Count of Workers with Respect to Race

```python
plt.figure(figsize=(7,5))
sns.countplot(x='race', data=df, palette='coolwarm')
plt.show()
```

# Work Industry Size

```python
plt.figure(figsize=(7,5))
sns.countplot(y='work_industry', data=df, palette='coolwarm')
plt.show()
```

# Admission Type Count

```python
plt.figure(figsize=(10,7))
sns.countplot(x='admission', data=df, palette='coolwarm')
plt.show()
```

# Workers Count with Respect to Major


```python
plt.figure(figsize=(10,7))
sns.countplot(x='major', data=df, palette='coolwarm')
plt.show()
```

# International Students (Female vs Male)

```python
plt.figure(figsize=(10,7))
sns.countplot(x='international', data=df, palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x='gender', data=df, palette='coolwarm')
plt.show()
```

```python
colors=['#b4cbf2','#ebbba3', '#d3eaf2', '#c7ecbe', '#ecbec2']
```

```python
crosstab=pd.crosstab(df['gender'], df['international'])
crosstab.plot(kind='bar', color=colors)
plt.xticks(rotation=0)
plt.show()
```

# Gender Ratio in Major

```python
crosstab=pd.crosstab(df['gender'], df['major'])
crosstab.plot(kind='bar', color=colors)
plt.xticks(rotation=0)
plt.show()
```

# Gender's by Race

```python
crosstab= pd.crosstab(df['gender'], df['race'])
crosstab.plot(kind='bar', color=colors)
plt.xticks(rotation=0)
plt.show()
```

# Gender Count in each Industry

```python
crosstab= pd.crosstab(df['gender'], df['work_industry'])
crosstab.plot(kind='bar', color=colors)
plt.xticks(rotation=0)
plt.show()
```

# GMAT Score Range of Work Industry

```python
plt.figure(figsize=(22,18))
sns.boxplot(x='work_industry', y='gmat', data=df, palette='coolwarm')
plt.xticks(rotation=90)
plt.show()
```

INDUSTRY WORK EXPERIENCE of WORKS

```python
plt.figure(figsize=(22,18))
sns.boxplot(x='work_industry', y='work_exp', data=df, palette='coolwarm')
plt.xticks(rotation=90)
plt.show()
```

GMAT SCORE RANGE BY GENDER

```python
plt.figure(figsize=(16,12))
sns.boxplot(x='race', y='gmat', data=df, palette='coolwarm')
plt.xticks(rotation=90)
plt.show()
```
